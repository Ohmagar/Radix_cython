```python
def radix(lst):
    def processor(arr):
        output_template = []
        arr_view = arr
        i = 0
        exponent = 0
        number_of_digits = 0
        current_index = 0
        current_element = 0
        start = 0
        stop = 0
        array_length = len(arr)
        position_within_buckets = 0
        output_array_position = 0
        count_of_digits = 0
        max_length = get_max_element_length(arr_view, array_length)
        output = [0] * array_length
        indexes = [0] * 10
        num_indexes = [i * array_length for i in range(max_length*10)]
        sorted_view = [0] * (array_length * max_length *10)
        output_view = output

        for i in range(array_length):
            exponent = 10
            current_element = arr_view[i]
            first_digit = 0
            for number_of_digits in range(max_length):
                if current_element < exponent:
                    break
                exponent *= 10
            first_digit = current_element//(exponent//10)
            sorted_view[num_indexes[number_of_digits*10+first_digit]] = current_element
            num_indexes[number_of_digits*10+first_digit] += 1
            
        
        for count_of_digits in range(max_length):
            c_o_d_10 = count_of_digits*10
            cod_1 = count_of_digits + 1
            for j in range(10):
                start = array_length * (c_o_d_10+j)
                stop = num_indexes[c_o_d_10+j]
                radix_int(sorted_view, start, stop, indexes, cod_1, 10, 1, stop - start, array_length)

        output_array_position = 0
        for count_of_digits in range(max_length):
            c_o_d_10 = count_of_digits*10
            for j in range(10):
                start = array_length * (c_o_d_10+j)
                stop = num_indexes[c_o_d_10+j]
                for position_within_buckets in range(start, stop):
                    output_view[output_array_position] = sorted_view[position_within_buckets]
                    output_array_position += 1

        return output

    def radix_int(unsorted, start, stop, indexes, max_length, p, p10, length, array_length):
        buckets = [0] * (length * 10)

        for j in range(max_length):
            for_loop(unsorted, start, stop, indexes, p, p10, length, buckets)
            p10 = p
            p *= 10

    def for_loop(unsorted_data, data_start, data_stop, indexes, p, p10, length, buckets):
        new_element_index = 0
        start = 0
        k_length = 0
        data_idx = data_start
        

        for k in range(10):
            indexes[k] = length * k
        for i in range(data_start, data_stop):
            new_element_index = (unsorted_data[i] % p) // p10
            buckets[indexes[new_element_index]] = unsorted_data[i]
            indexes[new_element_index] += 1

        for k in range(10):
            start = indexes[k]
            k_length = length * k
            for i in range(k_length, start):
                unsorted_data[data_idx] = buckets[i]
                data_idx += 1

        return unsorted_data

    def get_max_element_length(arr, length):
        i = 0
        p = 1
        current_max = arr[0]

        for i in range(length):
            if arr[i] > current_max:
                current_max = arr[i]

        i = 0
        while current_max // p != 0:
            i += 1
            p *= 10

        return i

    return processor(lst)
```
